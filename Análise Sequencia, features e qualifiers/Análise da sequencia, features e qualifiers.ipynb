```python
from Bio import Seq
from Bio import SeqIO
```

### Gene NLRP3

```python
#Lê o ficheiro no formato GenBank, e apresenta os resultados
#Variável N = NLRP3
N=SeqIO.read("sequence_nlrp3.gb", "genbank")
N
```

```python
print("**ID**")
print(N.id)
print("\n")
print("**Nome**")
print(N.name)
print("\n")
print("**Tamanho Sequência**")
print(len(N.seq))
print("\n")
print("**Descrição**")
print(N.description)
print("\n")
print("**Sequência Nucleotídica**")
print(N.seq)
print("\n")
```

### Features e Qualifiers


#### Features Gene NLRP3

```python
print("O número de features é:",len(N.features))
print("---------------------------------------")
print("Tipo e localização:")
for feature in N.features: 
    print(feature.type, feature.location)
```

```python
feat_cds = []
for f in range(len(N.features)):
    if N.features[f].type == "CDS":
        feat_cds.append(f)
conta_idx = len(feat_cds)
print(f"Quantidade de índices codificantes: {conta_idx}\n--------------------------------------")
print(f"Índice das sequências codificantes: {feat_cds}")
```
